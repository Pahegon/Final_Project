```python
df = pd.read_csv('FAO_csv/production_clean.csv')
df.Area.unique()
```

```python
df['Area'].replace({"United States of America":"United States"},inplace=True)
df.Area.unique()
```

```python

df.Area.unique()
```

```python
df = df.loc[df['Area'].isin(['Argentina', 'Brazil', 'Canada', 'China', 'France', 'Indonesia', 'India', 'Russian Federation', 'Ukraine', 'United States'])]
df.Area.unique()
```

```python
latitude = []

for row in df['Area']:
        if row == 'Argentina' :   latitude.append('-38.416097')
        elif row == 'Brazil' :   latitude.append('-14.235004')
        elif row == 'China' :   latitude.append('35.861660')
        elif row == 'France' :   latitude.append('46.227638')
        elif row == 'Indonesia' :   latitude.append('-0.789275')
        elif row == 'India' :   latitude.append('20.593684')
        elif row == 'Russian Federation' :   latitude.append('61.524010')
        elif row == 'Ukraine' :   latitude.append('48.379433')
        elif row == 'Canada' :   latitude.append('56.130366')
        elif row == 'United States' : latitude.append('37.090240')
        else: latitude.append('X')
```

```python
longitude = []

for row in df['Area']:
        if row == 'Argentina' :   longitude.append('-63.616672')
        elif row == 'Brazil' :   longitude.append('-51.925280')
        elif row == 'China' :   longitude.append('104.195397')
        elif row == 'France' :   longitude.append('2.213749')
        elif row == 'Indonesia' :   longitude.append('113.921327')
        elif row == 'India' :   longitude.append('78.962880')
        elif row == 'Russian Federation' :   longitude.append('105.318756')
        elif row == 'Ukraine' :   longitude.append('31.165580')
        elif row == 'Canada' :   longitude.append('-106.346771')
        elif row == 'United States' : longitude.append('-95.712891')

        else: longitude.append('Y')
```

```python
df['longitude'] = longitude
df['latitude'] = latitude
df
```

```python
df = df.drop(['Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'], axis=1)
df
```

```python
df.to_csv('Prod_Coord.csv')
```
